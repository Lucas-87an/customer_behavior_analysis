```python
path_dados = "/content/drive/MyDrive/AD/customer_shopping_behavior.csv"
path_saida_tabela = '/content/drive/MyDrive/AD/output/customer_shopping.csv'
```

```python
import pandas as pd
df_tabela = pd.read_csv(path_dados)
df_tabela
```

```python
df_tabela.info()
```

```python
df_tabela.describe(include='all')
```

```python
df_tabela.isnull().sum()
```

```python
df_tabela['Review Rating'] = df_tabela.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df_tabela.isnull().sum()
```

```python
df_tabela.columns = df_tabela.columns.str.lower()
df_tabela.columns = df_tabela.columns.str.replace(' ', '_')
df_tabela = df_tabela.rename(columns={'purchase_amount_(usd)': 'purchase_amount'})
```

```python
df_tabela.columns
```

```python
# create a column age_group
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df_tabela['age_group'] = pd.qcut(df_tabela['age'], q=4, labels = labels)
```

```python
df_tabela[['age', 'age_group']].head(10)
```

```python
# create column purchase_frequency_days

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df_tabela['purchase_frequency_days'] = df_tabela['frequency_of_purchases'].map(frequency_mapping)
```

```python
df_tabela[['purchase_frequency_days', 'frequency_of_purchases']].head(10)
```

```python
df_tabela[['discount_applied','promo_code_used']].head(10)
```

```python
(df_tabela['discount_applied'] == df_tabela['promo_code_used']).all()
```

```python
# Dropping promo code used column

df_tabela = df_tabela.drop('promo_code_used', axis=1)
```

```python
df_tabela.columns
```

```python
df_tabela.to_csv(path_saida_tabela, index=False)
```
